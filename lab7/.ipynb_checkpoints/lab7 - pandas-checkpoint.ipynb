```python
# Toturial adopted from https://github.com/jvns/pandas-cookbook
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

We're going to use a new dataset here, to demonstrate how to deal with larger datasets. This is a subset of the of 311 service requests from [NYC Open Data](https://nycopendata.socrata.com/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9). 

```python
complaints = pd.read_csv('311-service-requests.csv')
```

```python
complaints.head()
```

```python
complaints.shape
```

# 311 request by Complaint type

```python
complaint_counts = complaints['Complaint Type'].value_counts()
complaint_counts
```

```python
complaint_counts = complaint_counts.reset_index()
complaint_counts
```

```python
# Rename column 
complaint_counts.columns = ['complaint_type', 'counts']
complaint_counts.head()
```

### Plot Top 5  Complaints 

```python
type(complaint_counts)
```

```python
complaint_counts.head().plot(kind='bar', y='counts', x='complaint_type')
# Save the plot as file
plt.savefig('Top5Complaints.png')
```

# Datatime

```python
complaints.dtypes
```

```python
complaints['created_month'] = pd.to_datetime(complaints['Created Date']).dt.month
```

```python
complaints['created_hour'] = pd.to_datetime(complaints['Created Date']).dt.hour
```

```python
complaints.head()
```

# Groupby

```python
complaint_type_hour = complaints.groupby('Complaint Type')['created_hour'].mean()
```

```python
complaint_type_hour
```

```python
type(complaint_type_hour)
```

```python
# plot
plt.rcParams["figure.figsize"] = [50,9]
complaint_type_hour.plot()
plt.savefig('complaint_type_month.png')
```
